##Limpeza dos dados:

```python
print("Remover colunas desnecessárias")
df = df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'], errors='ignore')
```

```python
print("Remover linhas com valores nulos em Age e Embarked")
df = df.dropna(subset=['Age', 'Embarked'])
```

```python
print("Converter colunas Age e Fare para tipo numérico")
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce')
```

```python
print("Exibir informações gerais do dataset após a transfomação do tipo das variáveis:\n")
df.info()
```

```python
print("Verificando valores nulos apenas nas colunas 'Age' e 'Fare':\n")
df[['Age', 'Fare']].isnull().sum()
```

##Apareceram valores ausentes nas colunas Fare e Age após a transformação para valores numéricos.

**Solução**: Preencher valores nulos nessas colunas com a mediana para permitir a análise sem afetar e sem eliminar as linhas.

```python
print("Preenchendo valores nulos em 'Age' com a mediana...")
df['Age'] = df['Age'].fillna(df['Age'].median())

print("Preenchendo valores nulos em 'Fare' com a mediana...")
df['Fare'] = df['Fare'].fillna(df['Fare'].median())
```

##Após a limpeza dos dados:

```python
print("Informações gerais do dataset:\n")
df.info()
```

```python
print("Soma dos valores nulos:\n")
df.isnull().sum()
```

##8) O que foi alterado no dataset?
  - As colunas PassengerId, Name, Ticket e Cabin foram removidas. Elas foram consideradas desnecessárias para a análise direta ou continham muitos valores ausentes (Cabin).

  - Todas as linhas onde Age ou Embarked tinham valores nulos foram descartadas. Isso garante que não haja dados incompletos nessas variáveis cruciais.

  - As colunas Age e Fare foram convertidas para tipos numéricos (float64).

  - Após a conversão (devido a entradas que não puderam ser convertidas para números) foram preenchidos com a mediana de suas respectivas colunas. Isso preserva o maior número possível de dados sem introduzir distorções extremas.
